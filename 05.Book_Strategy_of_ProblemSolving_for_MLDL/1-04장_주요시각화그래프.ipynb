```python
import seaborn as sns 
import matplotlib.pyplot as plt
sns.set(style='darkgrid', palette="pastel")
sns.set(rc={'figure.figsize':(6, 3)})

titanic = sns.load_dataset('titanic')
print(titanic.shape)
titanic.head()
```

```python
#연령대별 히스토그램, 생존자수와 사망자수를 누적하여 표현 
sns.histplot(data=titanic, x='age', bins = 10, hue='alive', multiple='stack')
```

```python
#kdeplot: 커널밀도추정함수
sns.kdeplot(data=titanic, x='age', hue='alive', multiple='stack')
```

```python
#displot: histplot()과 kdeplot()을 동시에 
sns.displot(data=titanic, x='age', kde=True)
```

```python
#참고
sns.displot(data=titanic, x='age')
sns.displot(data=titanic, x='age', kind='kde')
```

```python
#rugplot: maginal distribution(주변분포), 단독보다 주로 다른 분포 차트와 병행하여 사용됨 
#x축위에 작은선분(rug)가 데이터의 밀집도를 표시해줌 
sns.kdeplot(data=titanic, x='age')
sns.rugplot(data=titanic, x='age') 
```
