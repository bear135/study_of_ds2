```python
## 필수 pkg's 및 한글폰트 사용 설정 
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt 
import seaborn as sns 
import warnings 
warnings.filterwarnings('ignore')

plt.rcParams['font.family'] = 'Malgun Gothic'
plt.rc('axes',unicode_minus=False)
#plt.rcParams["figure.figsize"] = (8,3)
pd.set_option('display.max_columns', None)
```

```python
df_temp = pd.read_csv('data/bdc_data_20230908.csv', encoding='cp949')

key_words = '리콜'      ## 포함되어야할 키워드 지정 
df_recall = df_temp[(df_temp['MEMO3'].str.contains(key_words, na=False))]  

print(df_recall.shape)
df_recall.head()

#df_recall.to_csv('output/df_recall.csv', index=False, encoding='utf-8-sig')    ## csv 파일 내보내기
```

```python
## 여러개의 단어 중 하나라도 포함하고 있는 것을 찾으려면 (or조건, *and 조건을 사용하려면 | 대신 &를 사용) 
key_words2 = '리콜|AS|엔진오일'
df_3words_or = df_temp[(df_temp['MEMO3'].str.contains(key_words2, na=False))]  

print(df_3words_or.shape)
df_3words_or.head()
```

```python
## "서비스 문의 여부 :O"로 기재된 것들만 찾기 
key_words3 = '서비스 문의 여부 :O'
df_3words_and = df_temp[(df_temp['MEMO3'].str.contains(key_words3, na=False))]  

print(df_3words_and.shape)
df_3words_and.head()
```
