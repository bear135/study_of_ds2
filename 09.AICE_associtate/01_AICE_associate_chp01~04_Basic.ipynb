##### Chp.04 - 기초 데이터 다루기 

```python
flight = pd.read_csv('data/Clean_Dataset.csv', encoding = "cp949")
flight.head()
```

```python
#df에서 데이터 선택 방법 : 1) 직접 컬럼명 지정 / 2) 범위지정 ~ flight[10:21] / 3) loc, iloc 사용 : 절대위치, 상대위치 
flight.loc[[102, 202, 302], ['airline', 'flight', 'source_city', 'price']]
```

```python
## iloc를 사용할 경우, 컬럼명을 직접 사용할 수 없다. 
flight.iloc[[102, 202, 302], [1,2,3,11]]
```

```python
# 데이터 삽입 : insert(행,(새로만들)열,값)
flight.insert(10, 'duration_2', flight['duration']*10)
flight.sample(3)
```

```python
# 데이터 삭제 : drop 사용. inplace 옵션으로 임시/영구 삭제 선택 
flight.drop('duration_2', axis=1, inplace=True)
flight.sample(3)
```

```python
# 컬럼명 변경과 정렬 : rename, sort_values 
flight = flight.rename(columns = {'Unnamed: 0':'ID', 'airline':'airline_name', 'source_city':'departure_city'})
flight = flight.sort_values(by='ID', ascending=True)
flight.head()
```

```python
# group by 
flight_tmp = flight[flight['class']=='Business']
flight_tmp.groupby(['airline_name', 'arrival_time']).min()
```

```python
# 결과의 원하는 부분의 데이터만 가져오기 
flight_tmp.groupby(['airline_name', 'arrival_time'])['price'].mean().loc[[('Air_India','Evening')]]
```

```python
# pivot, pivot_table : pivot은 중복값이 있으면 에러발생, 반면 pivot_table은 중복값 허용함 

## 샘플데이터 생성 
pivot_data = pd.DataFrame({'cust_id': ['cust_1', 'cust_1', 'cust_1', 'cust_2', 'cust_2', 'cust_2', 'cust_3', 'cust_3', 'cust_3'],
                  'prod_cd': ['p1', 'p2', 'p3', 'p1', 'p2', 'p3', 'p1', 'p2', 'p3'],
                  'grade' : ['A', 'A', 'A', 'A', 'A', 'A', 'B', 'B', 'B'],
                  'purch_amt': [30, 10, 0, 40, 15, 30, 0, 0, 10]})
pivot_data
```

```python
pivot_data.pivot_table(index='grade', columns='prod_cd', values='purch_amt', aggfunc=np.sum)
```

```python
pivot_data.pivot_table(index='grade', columns='prod_cd', values='purch_amt', aggfunc=np.mean)
```

```python
# index, column의 레벨 변경 
## 샘플데이터 
stack_data = pd.DataFrame({
    'Location': ['Seoul', 'Seoul', 'Seoul', 'kyounggi', 'kyounggi', 'Busan', 'Seoul', 'Seoul', 'Busan', 'kyounggi', 'kyounggi', 'kyounggi'],
    'Day': ['Mon', 'Tue', 'Wed', 'Mon', 'Tue', 'Mon', 'Thu', 'Fri', 'Tue', 'Wed', 'Thu', 'Fri'],
    'Ranfall': [100, 80, 1000, 200, 200, 100, 50, 100, 200, 100, 50, 100],
    'Rainfall_precipitation': [80, 70, 90, 10, 20, 30, 50, 90, 20, 80, 50, 10],
    'Temp' : [32, 27, 32, 31, 30, 28, 27, 25, 26, 33, 34, 31]})

stack_data
```

```python
## Location, Day를 인덱스로 변경 
new_stack_data = stack_data.set_index(['Location', 'Day'])
new_stack_data
```

```python
## unstak (index -> columns으로 변경)
new_stack_data.unstack(0)
```

```python
new_stack_data.unstack(1)
```

```python
## stack (반대로 columns -> index로 변경)
un_1 = new_stack_data.unstack(0)
un_1.stack(0)
```

```python
un_1.stack(1)
```
